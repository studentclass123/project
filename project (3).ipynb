```python
import pandas as pd
import plotly.express as px

# Sample data
data = {
    'total_bill': [16.99, 10.34, 21.01, 23.68, 24.59],
    'tip': [1.01, 1.66, 3.50, 3.31, 3.61],
    'sex': ['Female', 'Male', 'Male', 'Male', 'Female'],
    'day': ['Sun', 'Sun', 'Sun', 'Sun', 'Sun'],
    'time': ['Dinner', 'Dinner', 'Dinner', 'Dinner', 'Dinner'],
    'size': [2, 3, 3, 2, 4]
}

# Convert dictionary to DataFrame
df = pd.DataFrame(data)

# Check the DataFrame to ensure columns are correctly named
print(df.head())

# Create scatter plot
fig = px.scatter(
    data_frame=df,
    x='total_bill',
    y='tip',
    size='size',
    color='day',
    title='Scatter plot of Total Bill vs Tip',
    labels={'total_bill': 'Total Bill (in rupees)', 'tip': 'Tip (in rupees)'},
    hover_data=['sex', 'time']
)

# Show the plot
fig.show()

```

```python
figure = px.scatter(data_frame = data, x="total_bill",
                    y="tip", size="size", color= "day", trendline="ols")
figure.show()
```
